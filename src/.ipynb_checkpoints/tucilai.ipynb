```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Load dataset
data = pd.read_csv('../../data/data_train.csv')

# Statistik dasar
stats = data.describe()
print(stats)

duplikat = data.duplicated().sum()
print("Jumlah duplikat:", duplikat)

# Cek missing value
missing_value = data.isnull().sum()
print("Jumlah nilai yang hilang:")
print(missing_value)


kolom_numerik = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']
kolom_non_numerik = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']


# Histogram untuk kolom non-numerik
for col in kolom_non_numerik:
    data[col].value_counts().plot(kind='bar', title=col)
    plt.show()

# Distribusi data untuk kolom numerik
for col in kolom_numerik:
    plt.figure(figsize=(10,6))
    data[col].hist(bins=20, edgecolor='black')
    plt.title(col)
    plt.show()


# Korelasi dengan kolom target
korelasi = data.corr()['price_range'].sort_values(ascending=False)
print(korelasi)

# Korelasi chi-square antara kolom non numerik dengan kolom target
for col in kolom_numerik:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    batas_bawah = q1 - 1.5 * iqr
    batas_atas = q3 + 1.5 * iqr
    outlier = data[(data[col] < batas_bawah) | (data[col] > batas_atas)]
    print("Outlier pada", col, ":", len(outlier))

```
